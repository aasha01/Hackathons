## Cases by Date and Country

```python
country = covid.loc[:,covid.columns.isin(['Date','Case_Type','Cases','Country_Region'])]
country.head()
```

### Confirmed Cases Analysis

```python
confirmed = covid.loc[covid['Case_Type'] == 'Confirmed',covid.columns.isin(['Date','Case_Type','Cases','Country_Region'])]
```

```python
confirmed.set_index('Date',inplace = True)
```

```python
confirmed.head()
```

```python
# US Confirmed cases
case_by_country = confirmed.loc[confirmed['Country_Region'] == 'US',:]
case_by_country = case_by_country.groupby(by='Date').Cases.agg({np.sum})
case_by_country.head()
```

```python
#case_by_country.Country_Region.unique()
```

```python
plt.figure(figsize = (16,6))
#add a title
plt.title("Confirmed COVID-19 Cases in US")
sns.lineplot(data = case_by_country['sum'],label = 'Confirmed Cases')
plt.xlabel('Dates')
```

### Number of Deaths Analysis

```python
deaths = covid.loc[covid['Case_Type'] == 'Deaths',covid.columns.isin(['Date','Case_Type','Cases','Country_Region'])]
deaths.set_index('Date',inplace = True)
deaths.head()
```

```python
# US Deaths
case_by_country = deaths.loc[deaths['Country_Region'] == 'US',:]
case_by_country = case_by_country.groupby(by='Date').Cases.agg({np.sum})
case_by_country.head()
```

```python
plt.figure(figsize = (16,6))
#add a title
plt.title("Number of Deaths occurred due to COVID-19 in US")
sns.lineplot(data = case_by_country['sum'],label = 'Deaths')
plt.xlabel('Dates')
```

## Top 10 countries affected 

```python
confirmed_1 = covid.loc[covid['Case_Type'] == 'Confirmed',covid.columns.isin(['Date','Case_Type','Cases','Country_Region'])]
confirmed_1.head()
```

```python
confirmed_grp = confirmed_1.loc[confirmed_1['Date'] == '2020-05-02',confirmed_1.columns.isin(['Country_Region','Cases'])]
confirmed_grp = confirmed_grp.groupby(by='Country_Region').Cases.agg({np.sum})
```

```python
confirmed_grp.sort_values(by=['sum'],ascending=False, inplace=True)
top_10 = confirmed_grp.iloc[0:10,:]
top_10
```

```python
plt.figure(figsize = (16,6))
#add a title
plt.title("Top 10 countries affected by COVID-19 as of 2020-05-02")
g = sns.barplot(x = top_10.index,y = top_10['sum'])
plt.xlabel('Dates')

# Annotate every single Bar with its value, based on it's width    
for p in g.patches:
    g.annotate(p.get_height(), (p.get_x() + p.get_width() / 2., p.get_height()), ha = 'center', va = 'center', xytext = (0, 10), textcoords = 'offset points')
    
plt.show()
```

```python
sns.barplot(x= top_10['sum'],y=top_10.index, data=top_10)
```

## All countries Confirmed Cases

```python
confirmed_grp.sort_values(by=['sum'],ascending=False, inplace=True)
confirmed_grp.shape
confirmed_grp.tail()
```

```python
plt.figure(figsize = (10,100))
#add a title
plt.title("Countries affected by COVID-19 as of 2020-05-02")
ax = sns.barplot(x= confirmed_grp['sum'],y=confirmed_grp.index, data=confirmed_grp)
plt.xlabel('Dates')

# Annotate every single Bar with its value, based on it's width    
for p in ax.patches:
    width = p.get_width()
    ax.annotate(p.get_width(), (p.get_x() + p.get_width(), p.get_y() + 0.9),xytext=(5, 10), textcoords='offset points')

plt.show()
```

## Top 20 countries line plot

```python
AllCoun = covid.loc[covid['Case_Type'] == 'Confirmed',covid.columns.isin(['Date','Case_Type','Cases','Country_Region'])]
AllCoun.head()
AllCoun.drop(['Case_Type'],axis = 1, inplace=True)
#AllCoun.set_index(['Date'], append=True)
AllCoun.head()
```

```python
#To find to 20 countries
grp = AllCoun.sort_values(by='Cases',ascending=False)
grp_top = grp.groupby(by=['Date','Country_Region']).Cases.agg({np.sum})
```

```python
confirmed_1 = covid.loc[covid['Case_Type'] == 'Confirmed',covid.columns.isin(['Date','Case_Type','Cases','Country_Region'])]
confirmed_1.head()

confirmed_grp = confirmed_1.loc[confirmed_1['Date'] == '2020-05-02',confirmed_1.columns.isin(['Country_Region','Cases'])]
confirmed_grp = confirmed_grp.groupby(by='Country_Region').Cases.agg({np.sum})

confirmed_grp.sort_values(by=['sum'],ascending=False, inplace=True)
top_10 = confirmed_grp.iloc[0:20,:]
top_10
```

```python
country_list = top_10.index
country_list
```

```python
p = AllCoun.pivot_table(index='Date', columns='Country_Region', values='Cases', aggfunc=np.sum)
p.tail()
```

```python
top_20_Confirmed = p.loc[:,p.columns.isin(['US', 'Spain', 'Italy', 'United Kingdom', 'France', 'Germany', 'Turkey',
       'Russia', 'Brazil', 'Iran', 'China', 'Canada', 'Belgium', 'Peru',
       'Netherlands', 'India', 'Switzerland', 'Ecuador', 'Saudi Arabia',
       'Portugal'])]
top_20_Confirmed.head()
```

```python
top_6_Confirmed = p.loc[:,p.columns.isin(['US', 'Spain', 'Italy', 'United Kingdom', 'France'])]
```

```python
plt.figure(figsize = (16,6))
#add a title
plt.title("Top 20 countried confirmed with COVID-19 cases")
plt.xlabel('Dates')
sns.lineplot(data = top_20_Confirmed,dashes=False)
plt.show()
```

```python
top_19_Except_US = p.loc[:,p.columns.isin(['Spain', 'Italy', 'United Kingdom', 'France', 'Germany', 'Turkey',
       'Russia', 'Brazil', 'Iran', 'China', 'Canada', 'Belgium', 'Peru',
       'Netherlands', 'India', 'Switzerland', 'Ecuador', 'Saudi Arabia',
       'Portugal'])]
top_19_Except_US.head()
```

```python
plt.figure(figsize = (16,6))
#add a title
plt.title("Top 19 countried confirmed with COVID-19 cases Except US")
plt.xlabel('Dates')
sns.lineplot(data = top_19_Except_US,dashes=False)
plt.show()
```
