## Test Percentage analysis
We need only the columns having values for this analysis
Dropping na and assign it to new DataFrame

```python
TestPer = covid.loc[:,['People_Total_Tested_Count','Population_Count','Country_Region']]
TestPer.head()
```

```python
TestPer.Country_Region.unique()
```

```python
TestPer.isna().sum()
```

```python
TestPer.dropna(subset=['People_Total_Tested_Count'],inplace = True)
TestPer.isna().sum()
TestPer.dropna(subset=['Population_Count'],inplace = True)
TestPer.isna().sum()
```

```python
TestPer.Country_Region.unique()
```

```python
grp = TestPer.groupby(by='Country_Region').agg({np.sum})
grp
```

```python
grp['Tested Percentage'] = (grp['People_Total_Tested_Count']/grp['Population_Count'])*100
grp.head()
```
